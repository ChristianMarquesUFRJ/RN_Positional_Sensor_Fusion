```python
import plotly.express as px
import numpy as np
import plotly.graph_objects as go

pose = np.load("pose_real.npy")
gps = np.load("gps_real_1.npy")
mpu = np.load("mpu_1.npy")

fig1 = px.line(x = gps[0:150, 0]/np.linalg.norm( gps[0:150, 0]), y = gps[0:150, 1]/np.linalg.norm(gps[0:150, 1]) , markers=True)
```

```python

fig2 = px.line(x = pose[0:150, 1]/np.linalg.norm( pose[0:150, 1]), y = pose[0:150, 0]/np.linalg.norm(pose[0:150, 0]) , markers=True, color="r")
```
